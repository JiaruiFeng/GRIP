```python
import os

all_entity_list = []

for split in ['train', 'dev', 'test']:
    process_data = []
    with open(f"{split}.triples", "rb") as f:
        data = f.readlines()
    for line in data:
        line = line.decode("utf-8").split()
        if line[0][:7] != 'concept' or line[1][:7] != 'concept' or line[2][:7] != 'concept':
            print(line)
        else:
            process_data.append("\t".join([line[0],line[2],line[1]]))
            all_entity_list.append(line[0])
            all_entity_list.append(line[1])
    print(len(data), len(process_data), len(data) - len(process_data))
    
    file_name = 'valid' if split == 'dev' else split
    with open(f"{file_name}.txt", "w") as f:
        for i in range(len(process_data)):
          if i != len(process_data) - 1:
            f.write(process_data[i] + "\n")
          else:
            f.write(process_data[i])
```

```python
import json

entity_data = {}
for entity in set(all_entity_list):
    entity_data[entity] = {}

with open("entity2text.json", "w") as f:
    json.dump(entity_data, f, ensure_ascii=False, indent=4)
```
